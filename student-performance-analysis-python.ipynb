```python
### Business Question 1

Does studying longer improve examination scores?
```

```python
# Check for missing values
student_data.plot(
    x="study_hours",
    y="exam_score",
    kind="scatter"    
)
plt.title("Study Hours vs Exam Score")
plt.show()
```

```python
### showing the correlation coefficient for the study hours vs exam score
correlation = student_data ["study_hours"].corr(student_data["exam_score"])
print(f" Correlation coefficient:{correlation: .2f}")
```

Observation from the Correlation Coefficient:

The correlation coefficient between study hours and exam scores is 0.56, indicating a moderate positive relationship. This suggests that students who spend more time studying tend to achieve higher examination scores. However, the relationship is not strong enough to conclude that study hours alone determine academic performance. Other factors such as attendance, sleep, previous academic performance, and assignment completion are also likely to influence examination outcomes.

Observation:
Students who were successfully placed generally demonstrated stronger academic habits than students who were not placed. They also spent slightly less time using the internet and obtained substantially higher final examination scores (92.00 vs. 59.80). These findings suggest that consistent academic engagement is associated with successful placement.

## Phase 5: Key Findings

### Important takeaways

1. Study hours had the strongest relationship with examination performance.

The correlation coefficient between study hours and examination scores was 0.56, indicating a moderate positive relationship. Students who spent more time studying generally achieved higher examination scores.

2. Completing assignments was associated with better academic performance.

Assignment completion showed a positive relationship (0.39) with examination scores, suggesting that students who consistently completed assignments tended to perform better.

3. Previous academic performance was a useful predictor.

Students with higher previous scores generally obtained higher final examination scores (0.32), indicating that past academic success is associated with future performance.

4. Attendance had a positive but weaker relationship.

Attendance showed a weak positive correlation (0.22) with examination scores, suggesting that regular class attendance contributes to academic success.

5. Internet usage showed a weak negative relationship.

Internet usage had a correlation of -0.15, indicating that increased internet usage was slightly associated with lower examination scores.

6. Successfully placed students demonstrated stronger academic habits.

Students who were placed studied more hours, attended classes more regularly, completed more assignments, and achieved higher previous and final examination scores than students who were not placed.
